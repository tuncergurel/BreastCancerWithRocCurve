```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sklearn.model_selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(["Unnamed: 32"],axis=1,inplace=True)
```

```python
data.drop(["id"],axis=1, inplace=True)
```

```python
data.describe()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data.corr(),linewidths=.5,cmap="magma",annot=True,vmax=1, vmin=-1,center=0,mask=np.triu(np.ones_like(data.corr(), dtype=bool)))
plt.yticks(rotation = 0)
plt.suptitle("Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.loc[:,"radius_mean":"fractal_dimension_mean"].corr(),linewidths=1,cmap="magma",annot=True)
plt.yticks(rotation = 0)
plt.suptitle("Correlation")
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler(feature_range=(0,1))
```

```python
data = pd.DataFrame(mms.fit_transform(data),columns=data.columns)
```

```python
data.head()
```

```python
X = data.loc[:,data.columns != 'diagnosis']
y = data['diagnosis']
```

```python
X_Train, X_Test, y_train, y_test = train_test_split(X,y, test_size = 0.3 ,random_state = 20)
```

```python
X_Train.shape
```

```python
X_Test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(bootstrap=True,max_features=10, min_samples_leaf=30, min_samples_split=30)
```

```python
rfc.fit(X_Train,y_train)
```

```python
pred = rfc.predict(X_Test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix
```

```python
sns.heatmap(cm(y_test,pred),annot=True)
```

```python
from sklearn.metrics import roc_curve
rc = roc_curve
```

```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score
```

```python
f, t, _ = rc(y_test,pred)
```

```python
rauc = auc(y_test, pred)
```

```python
plt.plot(f,t)
plt.plot([0,1],[0,1], 'k--', linewidth = 2)
plt.show()
```

```python
plt.plot(f, t, label='AUC='+str(rauc))
plt.plot([0,1],[0,1], 'k--', linewidth = 2)
plt.legend(loc=4)
plt.ylabel("TP")
plt.xlabel("FP")
plt.show()
```
